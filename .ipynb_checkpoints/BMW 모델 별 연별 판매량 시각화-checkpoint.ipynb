```python
# 연간 브랜드별 판매량 꺾은 선 그래프 시각화

import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = 'Malgun Gothic'

# 차트 그리기
fig = plt.figure(figsize=(20, 12)) # 차트 생성 및 사이즈 설정
ax = fig.add_subplot(1,1,1) # subplot 생성

ax.plot(BMW_quentity, label='BMW', marker='o', color='#236E96') 
ax.plot(MB_quentity, label='MB', marker='o', color='#15B2D3') 
ax.plot(Audi_quentity, label='Audi', marker='o', color='#FFD700') 
ax.plot(Volkswagen_quentity, label='Volkswagen', marker='o', color='#F3872F') 
ax.plot(Mini_quentity, label='Mini', marker='o', color='#FF598F') 
ax.plot(Volvo_quentity, label='Volvo', marker='o', color='#707571') 

ax.set_title('연간 브랜드별 판매량 추이', fontsize=20) # 타이틀 설정
ax.set_ylabel('year', fontsize=14) # x축 설정
ax.set_xlabel('판매량', fontsize=14) # y축 설정

ax.legend(fontsize=12, loc='best') # 범례 설정 best로 해놓으면 가장 적절한 위치에 알아서 범례가 놓이게 됩니디

plt.show()
```
