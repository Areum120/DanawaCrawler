### BMW 차량 서비스센터& 차대차 교통사고 발생건수 시각화 ###

```python
#전체화면 버튼
plugins.Fullscreen(
    position='topright',
    title='전체화면',
    title_cancel='나가기',
    force_separate_button=True
).add_to(map)


#서울시구별영역표시
geo_path = "C:\\Users\\웍스컴바인\\Desktop\\work\\data_analysis\\CallAPI\\ds\\data/skorea_municipalities_geo_simple.json"
geo_str = json.load(open(geo_path, encoding="utf-8-sig"))


#지도시작점
map = folium.Map(location=[35.775301, 128.029343], zoom_start=6.5, tiles="Stamen Toner")

folium.TileLayer('CartoDB positron', name='Maptype1').add_to(map)
folium.TileLayer('Stamen Toner', name='Maptype2').add_to(map)
folium.TileLayer('Stamen Terrain', name='Maptype3').add_to(map)


bmwgroup = folium.FeatureGroup(name='합계')
map.add_child(bmwgroup)


bmw_g1 = folium.plugins.FeatureGroupSubGroup(bmwgroup, '서비스센터(red)')
map.add_child(bmw_g1)


# 1)버블차트

# 2)choropleth
map.choropleth(geo_data=geo_str, data=df2,
               columns=["자치구", "차대차사고"],
               key_on="feature.id", fill_color="BuPu",
               fill_opacity=0.7, line_opacity=0.3,
               name="차대차 사고밀도")


#서비스센터
for n in BMW_sc.index:
    folium.Marker(location=[BMW_sc["lat"][n],BMW_sc["lng"][n]],
                                tooltip=('<b>매장정보'+'<p><b> 매장이름 : '+BMW_sc["ShopName"][n] + '<p><b> 주소 : '
                                         +BMW_sc["address"][n]+
                                        '<p><b> 전화 : '+ BMW_sc["Tel"][n] + '<p><b> 딜러 : '+ BMW_sc["DealerName"][n]),
                      radius=10,
                      icon=folium.Icon(color = 'red',icon='wrench',prefix='fa')).add_to(bmw_g1)


map
```
